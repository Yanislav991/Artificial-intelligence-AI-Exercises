#### Florentine Families

1. One famous graph in social network analysis is the "Florentine families" graph. It is implemented by default in `networkx`. Each node represents a family, each edge - a marital relationship (i.e. someone from family A married someone from family B). Since marriage is mutual, the graph is undirected. Load it (and visualize it if you want). How many nodes does the shortest path from Pazzi to Peruzzi contain, including the starting and the final one?

```python
g = nx.florentine_families_graph()
```

```python
nx.draw(g)
```

```python
print(nx.shortest_path(g, source="Pazzi", target="Peruzzi"))
```

2. What is the name of the most important / famous family in this graph? 

```python
deg_cent = nx.degree_centrality(g)
print(deg_cent)
most_important_node = max(deg_cent.values(), key=float)

print(most_important_node)
```
